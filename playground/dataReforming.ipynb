```python
import pandas as pd 
```

```python
data_old = pd.read_csv("../COVID-19 Data/covid_19_clean_complete.csv")
data_old
```

```python
data_1 = data_old.drop([data_old.columns[0], data_old.columns[2], data_old.columns[3]], axis=1)
data_1
```

```python
data_1.nunique()
```

```python
data_new = data_1.groupby([data_1.columns[0], data_1.columns[1]], axis=0, sort=False).sum()
```

```python
data_new
```

```python
data_new.to_csv("../COVID-19 Data/data_reformed.csv")
```
