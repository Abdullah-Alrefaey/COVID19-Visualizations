# Loading Aggregated Data

```python
# Aggregated data from github
data_agg = pd.read_csv("../COVID-19 Data/data_aggregated.csv")
data_agg
```

```python
# Extracting needed column names 
date_index = data_agg.columns[0]
country = data_agg.columns[1]

countries = list(dict.fromkeys(data_agg[country]))
dates = list(dict.fromkeys(data_agg.loc[:, date_index]))
dates
```

```python
countries
```

## Sorting Each day of the records 

```python
# Sorting each date on it`s own 
sorteddata = data_agg[data_agg[date_index].str.contains(dates[0])].sort_values(by=list(data_agg.columns[2:]), ascending= False)
sorteddata
```

```python
dates.remove(dates[0])
dates
```

```python
for date in dates :
    srtdata = data_agg[data_agg[date_index].str.contains(date)].sort_values(by=list(data_agg.columns[2:]), ascending= False)
    print(srtdata)
    sorteddata = pd.concat([sorteddata, srtdata])
```

```python
sorteddata
```

```python
# sorteddata.to_csv("../COVID-19 Data/aggregated_sorted.csv")
```

```python
dates = list(dict.fromkeys(sorteddata.loc[:, date_index]))
sorteddata[sorteddata[date_index].str.contains(dates[0])].iloc[:15, : ]
```

## Removing Zero Confirmed Records For Visualization

```python
highVals = []

for row in range(len(sorteddata)):
    if sorteddata.iloc[row, :].Confirmed > 0 :
        highVals.append(row)
highVals
```

```python
newData = sorteddata.iloc[highVals, :]
```

```python
newData.to_csv("../COVID-19 Data/aggregated_sorted.csv")
```
